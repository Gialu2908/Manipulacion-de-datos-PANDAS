# **Missing values** ‚ùå

```
Image(filename='missing.png', width=300)
```

Detectar Missing Values:

```
supermercado_to_dic.isna()
```

A√±adir un nuevo producto con un dato faltante:

```
supermercado_to_dic = supermercado_to_dic.append({'Categoria' : 'Verduras' , 'Producto' : 'Tomate'} , ignore_index=True, )
supermercado_to_dic
```

```
supermercado_to_dic.isna().any()
```

```
supermercado_to_dic.isna().sum()
```

Eliminar valores faltantes:

```
supermercado_to_dic.dropna(inplace=True)
supermercado_to_dic
```

Reemplazar valores faltanes:

```
supermercado_to_dic = supermercado_to_dic.append({'Categoria' : 'Verduras' , 'Producto' : 'Cebolla', 'Precio' : 400} , ignore_index=True)
supermercado_to_dic
```

```
supermercado_to_dic.fillna(0)
```
